```python
total= 0 
RT = 0
mentions = 0


for chunk in tweets:
        
    total += chunk.shape[0]
    
    for tweet in chunk["text"]:
        if tweet[:1] == "@":
            mentions+=1
        if tweet[:2] == "RT":
            RT+=1
                     
    
```

```python
print("Total tweets: " + str(total))
print("Total RT: " + str(RT))
print("Total mentions: " +str(mentions))
print("Total normal tweets: " + str(total-RT-mentions))
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['Total', 'Mentions', 'RT', 'Normal Tweets']
students = [total,mentions,RT,total-mentions-RT]
ax.bar(langs,students)
plt.show()
```

```python
tweets = pd.read_json('data/tweets.json' , lines=True, chunksize=100000)
df_count = pd.Series(data=None)

for chunk in tweets:
    
    df = chunk[['screen_name']]
    df = df.groupby('screen_name')["screen_name"].count()
    
    df_count = df_count.add(df, fill_value=0)
    
    break
```

```python
df_1 = pd.DataFrame({"Total": df_count}, index=df_count.index).sort_values(by=['Total'], ascending=False)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = df_1.index[:10]
students = df_1.Total[:10]
ax.bar(langs,students)
plt.xticks(rotation=90)
plt.show()
```
